## Dropping missing values

In this project we are going to do:

* An study of the date where movies were added and realeased
* An study of the countries where movies were made 
* An inferencial analysis about the duration of the movies. 

Therefore, we will delete director and cast columns because they don't provide value for those objetives, and we will drop those missing values in the other columns because use missing values imputacion techniques is not good idea of all for columns such as: duration, date_added, and so on...

```python
df2 = df.drop(['director', 'cast'], axis = 1).copy()
```

```python
df2.dropna(inplace = True)
```

```python
df2.shape
```

## Data transformation

#### deleting 's' in show_id values

```python
df2['show_id'] = df2['show_id'].str.split('s').apply(lambda x: x[1]).copy()
df2['show_id'] = df2['show_id'].astype('int')
df2['show_id'].head()
```

#### separating month and year from date_added

```python
df2['month_added'] = df2['date_added'].str.split().apply(lambda x: x[0]).copy()
df2['year_added'] = df2['date_added'].str.split().apply(lambda x: x[-1]).astype('int').copy()
```

```python
df2.drop('date_added', axis = 1, inplace = True)
```

#### Let's extract all movies genres to make a column for each row within the data set

```python
movies_gen = df2[df2['type'] == 'Movie'].listed_in.values
shows_gen = df2[df2['type'] == 'TV Show'].listed_in.values
```

```python
ge_m = []
for i in range(len(movies_gen)):
    genres = movies_gen[i]
    ge_m+= genres.split(", ")
ge_m =list(set(ge_m))
```

```python
ge_s = []
for i in range(len(shows_gen)):
    genres = shows_gen[i]
    ge_s+= genres.split(", ")
ge_s =list(set(ge_s))
```

```python
movies_set = pd.DataFrame(index = df2[df2['type'] == 'Movie'].index, columns= [g+' Genre' for g in ge_m]).fillna(0)   
shows_set = pd.DataFrame(index = df2[df2['type'] == 'TV Show'].index, columns= [g+' Genre' for g in ge_s]).fillna(0)
```

```python
# df2.head()
```

```python
for i in movies_set.index:
    for x in [df2['listed_in'].loc[i]]:
        x2 = x.split(", ")
        for y in x2:
            movies_set[y+' Genre'].loc[i] = 1
```

```python
for i in shows_set.index:
    for x in [df2['listed_in'].loc[i]]:
        x2 = x.split(", ")
        for y in x2:
            shows_set[y+' Genre'].loc[i] = 1
```

```python
df2.drop('listed_in', axis = 1, inplace = True)
```

#### To display in a better way this table let's create ranges of years

```python
#df2['release_year'].min()
years = pd.cut(df2['release_year'], bins =8, right = False)
```

## Let's to analyze firts by movies genres and then by tv shows genres.

```python
movies = df2[df2['type'] == 'Movie'].copy()
movies = pd.concat([movies, movies_set], axis = 1)
years_m = years[df2['type'] == 'Movie'].copy()
tv_shows = df2[df2['type']  == 'TV Show'].copy()
tv_shows = pd.concat([tv_shows, shows_set], axis = 1)
years_s = years[df2['type'] == 'TV Show'].copy()
```

## We are going to graph total movies and tv shows released per decade

```python
movies.groupby(years_m)['type'].value_counts().plot.bar(color = color)
```

## Now, let's get the percentage of movies genres and tv shows genres released per decade 

```python
movies2 = movies.groupby(years_m)[[g + ' Genre' for g in ge_m]].sum()
movies2['total'] = movies2.sum(axis = 1)
tv_shows2 = tv_shows.groupby(years_s)[[g + ' Genre' for g in ge_s]].sum()
tv_shows2['total'] = tv_shows2.sum(axis = 1)
```

```python
movies2 = movies2.apply(lambda x: x/movies2.total)
tv_shows2 = tv_shows2.apply(lambda x: x/tv_shows2.total)
movies2.drop('total', inplace = True, axis = 1)
tv_shows2.drop('total', inplace = True, axis = 1)
tv_shows2 = tv_shows2.fillna(0)
```

# Analysis of duration of movies and tv shows

```python
duration_m = df2[df2['type'] == 'Movie']
duration_s = df2[df2['type'] == 'TV Show']
```

### Preparing duration feature for analysis

#### Deleting "minutes" and "season" in duration column

```python
duration_m.duration = duration_m.duration.str.split().apply(lambda x: x[0]).astype('int')
duration_s.duration = duration_s.duration.str.split().apply(lambda x: x[0]).astype('int')
```

## Let's grapg data distribution

```python
plt.figure(figsize=(10, 5))
plt.subplot(121)
sns.distplot(duration_m.duration, kde = True)
plt.title('Distribution of duration of movies')
plt.subplot(122)
sns.distplot(duration_s.duration, kde = True)
plt.title('Distrubution of duration of tv shows')
plt.show()
```

```python
print(duration_m.duration.mean())
print(duration_s.duration.mean())
```

## Aplicacion of inferential statistic

### Suppose we doesn't have access to entire poblation but to a sample. We are going to infer the confidence intervals for the mean duration of movies and tv shows for the entire poblation.

For that, we are going to use bootstrap method, to create resamples and don't have the need of assumptions about the population.

### Drowing sample

```python
sample_m = duration_m.duration.sample(100).values
sample_s = duration_s.duration.sample(100).values
```

## Drawing resamples for the original sample

```python
from sklearn.utils import resample

boot_m = [resample(sample_m, n_samples = len(sample_m), replace = True).mean() for _ in range(1000)]
boot_s = [resample(sample_s, n_samples = len(sample_s), replace = True).mean() for _ in range(1000)]
```

## Lest's graph the bootstrap distribution for the mean

```python
from scipy.stats import norm
plt.figure(figsize=(20, 5))
plt.subplot(121)
sns.distplot(boot_m, fit = norm)
plt.title('Bootstrap distribution for movies duration mean')
plt.subplot(122)
sns.distplot(boot_s, fit = norm)
plt.title('Bootstrap distribution for tv shows duration mean')
plt.show()
```

## let's get the confidence intervals of 95 percent using percentile bootstrap intervals

```python
quantile_m = np.quantile(a = boot_m, q=[0.025, 0.975])
quantile_s = np.quantile(a = boot_s, q = [0.025, 0.975])
```

```python

print(f'The average duration of movies on netflix is between {round(quantile_m[0], 2)} minutes and {round(quantile_m[1], 2)} min with 95% of confidence')
print(f'The average duration of tv shows on netflix is between {round(quantile_s[0], 2)} seasons and {round(quantile_s[1], 2)} seasons with 95% of confidence')
```

## Conclusion

As we can see, the $\mu$ for movies duration and tv shows duration are within their confidence intervals calculted with bootstrap method.
